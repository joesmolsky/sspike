# Warren 2020 models

```python
from sspike import pnut, beer
from sspike.snowball import Snowball
from sspike.targets import Target
```

```python
model = 'Warren_2020'
progenitor = {'mass': 13.0,
              'stir': 1.23}
transformation = 'NoTransformation'
distance = 5
```

```python
sb = Snowball(model, progenitor, transformation, distance)
target = Target('kamland')
snowflakes = pnut.snowglobes_events(sb, target)
sspikes = pnut.sspike_events(sb, target)
```

```python
nc_chans = target.nc_channels
beer.draw(sspikes[-1], channels=nc_chans)
```

```python
smear_path = snowflakes[1]
snow_chans = target.snow_channels
beer.draw(smear_path, channels=snow_chans)
```

```python
basic_chans = target.basic_channels
beer.draw(sspikes[0], channels=basic_chans)
```

```python
tab = beer.tab(sb)
```

```python
unsmeared_path = snowflakes[0]
snow_chans = target.snow_channels[:7]
beer.draw(unsmeared_path, channels=snow_chans)
```
