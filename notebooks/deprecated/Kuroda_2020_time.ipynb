# Supernova event rates vs. time

```python
from sspike import pnut, beer
from sspike.snowball import Snowball
from sspike.detectors import Detector
```

```python
model = 'Kuroda_2020'
progenitor = {'omega': '10',
              'B0': '13'}
transformation = 'NoTransformation'
distance = 0.6
t_bins = 100
```

```python
bliz = Snowball(model, progenitor, transformation, distance, t_bins)
detector = Detector('kamland')
sno = pnut.time_events(bliz.tarball, detector.name)
```

```python
times = beer.bin_times(bliz)
plot_type = '_unsmeared_weighted'
beer.series_plot(bliz, sno, times, plot_type)
```
